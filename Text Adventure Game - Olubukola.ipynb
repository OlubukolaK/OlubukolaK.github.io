Lagertha gathered a small group of loyal supporters, and together they devised a daring plan to infiltrate Jarl Borg's heavily fortified refuge. They carefully planned their every move, knowing that Jarl Borg's refuge was renowned for its impenetrable defences.
Under the cover of darkness, Lagertha and her allies stealthily approached the refuge. They moved through the dense forest, evading watchful guards and navigating treacherous paths. Their hearts pounded with anticipation and fear.
Finally, they reached Jarl's inner refuge, a grand hall where he held court. The air crackled with tension as Lagertha emerged from the shadows. Her presence commanded attention, striking fear into the hearts of Jarl Borg's followers.
A fierce confrontation ensued, with Lagertha and her allies facing off against Jarl Borg's formidable warriors. The hall echoed with the clash of steel as swords met and shields parried. The air was thick with the scent of sweat and blood.
Despite the odds stacked against her, Lagertha fought with unwavering determination. Her swordsmanship was unmatched, her movements swift and precise. She cut through Jarl Borg's men with a determination fuelled by the years of injustice she had endured.

The game 
In this game, the goal is to help Lagertha kill Jarl Borg.

```python
import time

def welcome_message():
    print("Welcome to the Viking Adventure Game!")
    print("You are Lagertha, a fierce shieldmaiden seeking revenge against Jarl Borg.")
    print("Your journey will take you through treacherous forests, guarded strongholds, and fierce battles.")
    print("Will you emerge victorious and claim your vengeance?")
    time.sleep(2)

def stage_one():
    print("You stand at the edge of Jarl Borg's stronghold, its imposing walls towering above you.")
    print("The guards are alert, their eyes scanning the darkness.")
    print("Do you attempt to sneak past the guards (1) or challenge them openly (2)?")

    user_choice = input("Enter your choice: ")

    if user_choice == "1":
        sneak_past_guards()
    elif user_choice == "2":
        challenge_guards()
    else:
        print("Invalid choice. Please enter either 1 or 2.")
        stage_one()

def sneak_past_guards():
    print("You carefully navigate the shadows, avoiding the watchful eyes of the guards.")
    print("You manage to slip past unnoticed and enter the stronghold undetected.")
    print("You proceed to the inner sanctum, where Jarl Borg awaits.")
    stage_two()

def challenge_guards():
    print("You draw your sword and charge towards the guards, their weapons glinting in the moonlight.")
    print("A fierce battle ensues, the clash of steel echoing through the night.")
    print("You fight with skill and determination, but the guards are numerous and well-trained.")

    user_choice = input("Do you continue fighting (1) or retreat to the shadows (2)?")

    if user_choice == "1":
        continue_fighting()
    elif user_choice == "2":
        retreat_to_shadows()
    else:
        print("Invalid choice. Please enter either 1 or 2.")
        challenge_guards()

def continue_fighting():
    print("You press on, your sword slicing through the ranks of Jarl Borg's warriors.")
    print("But the battle takes its toll, and you are eventually overwhelmed.")
    fail_function()

def retreat_to_shadows():
    print("You break away from the fight and seek refuge in the darkness.")
    print("You manage to escape the stronghold and regroup with your allies.")
    print("You decide to formulate a new plan to infiltrate the stronghold.")
    stage_one()

def stage_two():
    print("You enter the inner sanctum, a grand hall filled with Jarl Borg's followers.")
    print("Jarl Borg sits upon a throne, his eyes fixed upon you.")
    print("Do you confront Jarl Borg directly (1) or use stealth to approach him (2)?")

    user_choice = input("Enter your choice: ")

    if user_choice == "1":
        confront_jarl_borg()
    elif user_choice == "2":
        use_stealth()
    else:
        print("Invalid choice. Please enter either 1 or 2.")
        stage_two()

def confront_jarl_borg():
    print("You boldly approach Jarl Borg, your sword drawn and ready.")
    print("A fierce duel ensues, the fate of Hedeby hanging in the balance.")
    print("You fight with unwavering determination, fueled by the years of injustice you have endured.")

    user_choice = input("Do you attack with power (1) or precision (2)?")

    if user_choice == "1":
        attack_with_power()
    elif user_choice == "2":
        attack_with_precision()
    else:
        print("Invalid choice. Please enter either 1 or 2.")
        confront_jarl_borg()

def attack_with_power():
    print("You unleash a series of powerful blows, shattering Jarl Borg's defenses.")
    print("He falls to the ground, his lifeless body at your feet.")
    win_function()

def attack_with_precision():
    print("You strike with swift, precise movements, exploiting Jarl Borg's weaknesses.")
    print("He is disarmed and defeated, kneeling before you in surrender.")
    print("You have avenged the wrongs he has inflicted upon Hedeby.")
    win_function()

def use_stealth():
    print("You move silently through the hall, avoiding the attention of Jarl Borg's followers.")
    print("You manage to catch Jarl Borg off guard, your blade poised at his throat.")
    print("He is helpless, his life in your hands.")

    user_choice = input("Do you strike him down (1) or spare his life (2)?")

    if user_choice == "1":
        strike_jarl_borg()
    elif user_choice == "2":
        spare_jarl_borg()
    else:
        print("Invalid choice. Please enter either 1 or 2.")
        use_stealth()

def strike_jarl_borg():
    print("You plunge your sword into Jarl Borg's heart, ending his reign of terror.")
    print("The hall erupts in chaos as Jarl Borg's followers scramble to escape.")
    print("You have avenged the wrongs he has inflicted upon Hedeby.")
    win_function()

def spare_jarl_borg():
    print("You lower your sword, deciding to show mercy.")
    print("Jarl Borg is stunned and speechless, his arrogance shattered.")
    print("You banish him from Hedeby, warning him never to return.")
    print("You have restored justice to Hedeby and defeated your enemies without bloodshed.")
    win_function()

def win_function():
    print("Congratulations! You have emerged victorious and achieved your vengeance.")
    print("Your bravery and determination have earned you the respect and admiration of all who know you.")
    print("You are a true warrior queen, an inspiration to women everywhere.")
    time.sleep(3)
    print("The End.")

def fail_function():
    print("Jarl Borg's warriors overwhelm you, their blades finding their mark.")
    print("You fall to the ground, your vision fading as darkness consumes you.")
    print("Jarl Borg's reign of terror continues, his grip on Hedeby unwavering.")
    print("You have failed to achieve your vengeance, but your spirit remains unbroken.")
    time.sleep(3)
    print("The End.")

welcome_message()
stage_one()

```
